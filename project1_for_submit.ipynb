```python
!sudo apt-get install -y fonts-nanum
!sudo fc-cache -fv
!rm ~/.cache/matplotlib -rf

import pandas as pd
import requests

# GitHub에 업로드된 파일의 URL 설정
subway_data_url = "https://raw.githubusercontent.com/stack9798/project1/main/subway_data.csv"
bus_user_data_url = "https://raw.githubusercontent.com/stack9798/project1/main/bus_user.csv"
bus_station_data_url = "https://raw.githubusercontent.com/stack9798/project1/main/bus_station.csv"

# 파일 다운로드 함수
def download_file(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded {filename} from {url}")
    else:
        print(f"Failed to download {filename} from {url}. Status code: {response.status_code}")

# 파일 다운로드
download_file(subway_data_url, "subway_data.csv")
download_file(bus_user_data_url, "bus_user.csv")
download_file(bus_station_data_url, "bus_station.csv")

# CSV 파일 경로 설정
subway_file_path = "subway_data.csv"
bus_user_file_path = "bus_user.csv"
bus_station_file_path = "bus_station.csv"

# CSV 파일 불러오기
subway_data = pd.read_csv(subway_file_path)
bus_user_data = pd.read_csv(bus_user_file_path)
bus_station_data = pd.read_csv(bus_station_file_path)


import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'NanumBarunGothic' # 나눔바른고딕 적용하기

# 지하철 혼잡도 데이터 관련 표 만들기
# 조건에 맞는 데이터 추출하기
subway_to_Sangdo_weekday = subway_data[(subway_data['호선'] == 7) & (subway_data['출발역'] == "숭실대입구") & (subway_data['상하구분'] == '상선') & (subway_data['요일구분'] == '평일')]
subway_to_Sangdo_weekend = subway_data[(subway_data['호선'] == 7) & (subway_data['출발역'] == "숭실대입구") & (subway_data['상하구분'] == '상선') & ((subway_data['요일구분'] == '토요일') | (subway_data['요일구분'] == '일요일'))]
subway_to_Namsung_weekday = subway_data[(subway_data['호선'] == 7) & (subway_data['출발역'] == "숭실대입구") & (subway_data['상하구분'] == '하선') & (subway_data['요일구분'] == '평일')]
subway_to_Namsung_weekend = subway_data[(subway_data['호선'] == 7) & (subway_data['출발역'] == "숭실대입구") & (subway_data['상하구분'] == '하선') & ((subway_data['요일구분'] == '토요일') | (subway_data['요일구분'] == '일요일'))]

# 테두리가 있는 HTML 표로 변환
table_Sonngsill_Sangdo_weekday = subway_to_Sangdo_weekday.style.set_properties(**{'border': '1px solid white'}).to_html()
table_Sonngsill_Sangdo_weekend = subway_to_Sangdo_weekend.style.set_properties(**{'border': '1px solid white'}).to_html()
table_Sonngsill_Namsung_weekday = subway_to_Namsung_weekday.style.set_properties(**{'border': '1px solid white'}).to_html()
table_Sonngsill_Namsung_weekend = subway_to_Namsung_weekend.style.set_properties(**{'border': '1px solid white'}).to_html()

# HTML 표 출력
from IPython.display import HTML, display

display(HTML(table_Sonngsill_Sangdo_weekday))
display(HTML(table_Sonngsill_Sangdo_weekend))
display(HTML(table_Sonngsill_Namsung_weekday))
display(HTML(table_Sonngsill_Namsung_weekend))

# 지하철 혼잡도 데이터 선그래프 시각화
# 첫 번째 그래프 그리기
plt.figure(figsize=(20, 6))

# x축과 y축 데이터 설정
x = subway_to_Sangdo_weekday.columns[6:]  # 7열부터 모든 열의 열 이름을 x축으로 사용
y = subway_to_Sangdo_weekday.iloc[0, 6:]  # 2행의 7열부터 모든 열 y축으로 사용

# 그래프 그리기
plt.plot(x, y, marker='o', linestyle='-')

# 축 및 타이틀 설정
plt.xlabel('시간대')
plt.ylabel('승객 비율 (%)')
plt.title('숭실대입구 → 상도 방면 평일 승객 비율')

# 그래프 출력
plt.xticks(rotation=45)  # x축 라벨 회전
plt.grid(True)  # 그리드 표시
plt.tight_layout()
plt.show()

# 두번째 그래프
plt.figure(figsize=(20, 6))
x = subway_to_Sangdo_weekend.columns[6:]  # 7열부터 모든 열의 열 이름을 x축으로 사용
y_sat = subway_to_Sangdo_weekend.iloc[0, 6:]  # 1행의 7열부터 모든 열, y축으로 사용

# 일요일 데이터 설정
y_sun = subway_to_Sangdo_weekend.iloc[1, 6:]  # 2행의 7열부터 모든 열, y축으로 사용

# 그래프 그리기
plt.plot(x, y_sat, marker='o', linestyle='-', label='토요일')
plt.plot(x, y_sun, marker='o', linestyle='-', label='일요일')

# 축 및 타이틀 설정
plt.xlabel('시간대')
plt.ylabel('승객 비율 (%)')
plt.title('숭실대입구 → 상도 방면 주말 승객 비율')
plt.legend()  # 범례 표시

# 그래프 출력
plt.xticks(rotation=45)  # x축 라벨 회전
plt.grid(True)  # 그리드 표시
plt.tight_layout()
plt.show()

# 세번째 그래프 (남성 방면 평일)
plt.figure(figsize=(20, 6))

# x축과 y축 데이터 설정
x = subway_to_Namsung_weekday.columns[6:]  # 7열부터 모든 열의 열 이름을 x축으로 사용
y = subway_to_Namsung_weekday.iloc[0, 6:]  # 2행의 7열부터 모든 열 y축으로 사용

# 그래프 그리기
plt.plot(x, y, marker='o', linestyle='-')

# 축 및 타이틀 설정
plt.xlabel('시간대')
plt.ylabel('승객 비율 (%)')
plt.title('숭실대입구 → 남성 방면 평일 승객 비율')

# 그래프 출력
plt.xticks(rotation=45)  # x축 라벨 회전
plt.grid(True)  # 그리드 표시
plt.tight_layout()
plt.show()

# 네번째 그래프 (남성 방면 주말)
plt.figure(figsize=(20, 6))
x = subway_to_Namsung_weekend.columns[6:]  # 7열부터 모든 열의 열 이름을 x축으로 사용
y_sat = subway_to_Namsung_weekend.iloc[0, 6:]  # 1행의 7열부터 모든 열, y축으로 사용

# 일요일 데이터 설정
y_sun = subway_to_Namsung_weekend.iloc[1, 6:]  # 2행의 7열부터 모든 열, y축으로 사용

# 그래프 그리기
plt.plot(x, y_sat, marker='o', linestyle='-', label='토요일')
plt.plot(x, y_sun, marker='o', linestyle='-', label='일요일')

# 축 및 타이틀 설정
plt.xlabel('시간대')
plt.ylabel('승객 비율 (%)')
plt.title('숭실대입구 → 남성 방면 주말 승객 비율')
plt.legend()  # 범례 표시

# 그래프 출력
plt.xticks(rotation=45)  # x축 라벨 회전
plt.grid(True)  # 그리드 표시
plt.tight_layout()
plt.show()

# 버스 혼잡도(재차 인원) 데이터 표로 추출 및 그래프로 시각화
from IPython.display import display

# 필요한 정류장명 필터링
filtered_data = bus_user_data[bus_user_data['정류장명'].str.contains('숭실대정문앞|숭실대별관앞|숭실대후문|숭실대입구역')]

# 같은 노선 번호와 정류장명에 대해 정류장 순번이 작은 것만 추출
filtered_unique_data = filtered_data.loc[filtered_data.groupby(['노선', '정류장명'])['정류장순번'].idxmin()]

# 필터링된 데이터프레임 출력
display(filtered_unique_data)

# 시간대 컬럼들
time_columns = ['04시', '05시', '06시', '07시', '08시', '09시', '10시', '11시', '12시',
                '13시', '14시', '15시', '16시', '17시', '18시', '19시', '20시', '21시',
                '22시', '23시', '00시', '01시', '02시', '03시']

# 정류장별로 데이터를 나눔
bus_stops = {
    '숭실대입구역': '20810',
    '숭실대정문앞': '20169',
    '숭실대별관앞': '20172',
    '숭실대후문': '20535'
}

# 그래프 그리기
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 15), sharex=True, sharey=True)
axes = axes.flatten()

for i, (bus_stop, stop_code) in enumerate(bus_stops.items()):
    if i >= len(axes):
        break  # 플롯 개수를 초과하지 않도록 함

    stop_data = filtered_unique_data[filtered_unique_data['정류장명'] == bus_stop]

    for _, row in stop_data.iterrows():
        axes[i].plot(time_columns, row[time_columns], label=f"{row['노선']}번 노선")

    axes[i].set_title(f"{bus_stop} ({stop_code})")
    axes[i].set_xticks(range(len(time_columns)))
    axes[i].set_xticklabels(time_columns, rotation=45)
    axes[i].set_ylabel('재차인원')
    axes[i].legend()

# 공통 X, Y 라벨 설정
fig.text(0.5, 0.04, '시간', ha='center')
fig.text(0.04, 0.5, '재차인원', va='center', rotation='vertical')

plt.tight_layout()
plt.show()

# 응답 프로그램 코드

import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# CSV 파일 경로 설정
subway_file_path = "subway_data.csv"
bus_user_file_path = "bus_user.csv"

# CSV 파일 불러오기
subway_data = pd.read_csv(subway_file_path)
bus_user_data = pd.read_csv(bus_user_file_path)

# 시간을 가장 가까운 시간대로 매핑하는 함수 (버스)
def map_time_to_nearest_column_bus(time):
    hour, minute = map(int, time.split(':'))
    if minute >= 30:
        hour += 1
    return f"{hour:02d}시"

# 시간을 가장 가까운 30분 단위로 매핑하는 함수 (지하철)
def map_time_to_nearest_column_subway(time):
    hour, minute = map(int, time.split(':'))
    if minute < 15:
        return f"{hour}시00분"
    elif minute < 45:
        return f"{hour}시30분"
    else:
        return f"{hour + 1}시00분"

# 혼잡도 계산 함수 (지하철)
def get_congestion_level_subway(direction, day, time):
    time_column = map_time_to_nearest_column_subway(time)

    hour, minute = map(int, time.split(':'))

    if day == "평일" and ((hour == 1 and minute > 0) or (1 < hour < 5) or (hour == 5 and minute < 30)):
        return "운행하지 않는 시간대입니다."
    elif day in ["토요일", "일요일"] and ((hour >= 23 or (hour < 5)) or (hour == 5 and minute < 30)):
        return "운행하지 않는 시간대입니다."

    if direction == "상도":
        data = subway_data[(subway_data['호선'] == 7) &
                           (subway_data['출발역'] == "숭실대입구") &
                           (subway_data['상하구분'] == '상선')]
    else:
        data = subway_data[(subway_data['호선'] == 7) &
                           (subway_data['출발역'] == "숭실대입구") &
                           (subway_data['상하구분'] == '하선')]

    if day == "평일":
        data = data[data['요일구분'] == '평일']
    else:
        data = data[data['요일구분'].isin(['토요일', '일요일'])]

    if time_column not in data.columns:
        return "해당 시간대의 데이터가 없습니다."

    passenger_ratio = data[time_column].mean()
    if passenger_ratio > 60:
        return "매우 혼잡"
    elif passenger_ratio > 34:
        return "혼잡"
    elif passenger_ratio > 20:
        return "쾌적"
    else:
        return "매우 쾌적"

# 혼잡도 계산 함수 (버스)
def get_congestion_level_bus(stop, line, time):
    time_column = map_time_to_nearest_column_bus(time)

    data = bus_user_data[(bus_user_data['정류장명'] == stop) &
                         (bus_user_data['노선'] == line)]

    if time_column not in data.columns:
        return "해당 시간대의 데이터가 없습니다."

    passengers = data[time_column].mean()

    if passengers > 25:
        return "매우 혼잡"
    elif passengers > 20:
        return "혼잡"
    elif passengers > 15:
        return "보통"
    elif passengers > 10:
        return "쾌적"
    else:
        return "매우 쾌적"

# 시간 형식 검증 함수
def validate_time_format(time):
    try:
        hour, minute = map(int, time.split(':'))
        if 0 <= hour <= 23 and 0 <= minute <= 59:
            return True
        else:
            return False
    except ValueError:
        return False

# GUI 요소 정의
mode_label = widgets.Label("지하철을 이용하시겠습니까, 버스를 이용하시겠습니까?")
mode_dropdown = widgets.Dropdown(options=['지하철', '버스'], value='지하철')

# 지하철 관련 요소
direction_label = widgets.Label("방면을 선택하세요 (상도/남성):")
direction_dropdown = widgets.Dropdown(options=['상도', '남성'], value='상도')

day_label = widgets.Label("요일을 선택하세요 (평일/토요일/일요일):")
day_dropdown = widgets.Dropdown(options=['평일', '토요일', '일요일'], value='평일')

# 버스 관련 요소
bus_stop_label = widgets.Label("정류장을 선택하세요:")
bus_stop_dropdown = widgets.Dropdown(options=['숭실대입구역', '숭실대정문앞', '숭실대별관앞', '숭실대후문'], value='숭실대입구역')

line_label = widgets.Label("이용할 노선을 선택하세요:")
line_dropdown = widgets.Dropdown()

time_label = widgets.Label("이용할 시간을 입력하세요 (24시 기준, 예: 15:30):")
time_text = widgets.Text(value='15:30')

result_label = widgets.Label("혼잡도: ")
result_text = widgets.Text(value='', disabled=True)

# 노선 드롭다운 업데이트
def update_line_dropdown(*args):
    stop = bus_stop_dropdown.value
    available_lines = bus_user_data[bus_user_data['정류장명'] == stop]['노선'].unique()
    line_dropdown.options = available_lines
bus_stop_dropdown.observe(update_line_dropdown, 'value')
update_line_dropdown()

# 응답 함수
def on_submit(button):
    mode = mode_dropdown.value
    time = time_text.value
    if not validate_time_format(time):
        result_text.value = "시간 형식이 잘못되었습니다. 다시 입력해주세요."
        return
    if mode == "지하철":
        direction = direction_dropdown.value
        day = day_dropdown.value
        level = get_congestion_level_subway(direction, day, time)
    elif mode == "버스":
        stop = bus_stop_dropdown.value
        line = line_dropdown.value
        level = get_congestion_level_bus(stop, line, time)
    result_text.value = f"혼잡도: {level}"

submit_button = widgets.Button(description="혼잡도 확인")
submit_button.on_click(on_submit)

# 모드에 따른 위젯 표시
def update_ui(mode):
    if mode == '지하철':
        return widgets.VBox([
            mode_label, mode_dropdown,
            direction_label, direction_dropdown,
            day_label, day_dropdown,
            time_label, time_text,
            submit_button, result_label, result_text
        ])
    elif mode == '버스':
        return widgets.VBox([
            mode_label, mode_dropdown,
            bus_stop_label, bus_stop_dropdown,
            line_label, line_dropdown,
            time_label, time_text,
            submit_button, result_label, result_text
        ])

# 모드 변경 시 UI 업데이트
def on_mode_change(change):
    ui.children = [update_ui(change['new'])]

mode_dropdown.observe(on_mode_change, names='value')

# 초기 UI 설정
ui = widgets.VBox([update_ui(mode_dropdown.value)])
display(ui)

# 버스 정류장 위도, 경도 불러오기

# 파일 경로 설정
bus_station_file_path = "bus_station.csv"

# 파일 읽기
bus_station_data = pd.read_csv(bus_station_file_path)

# 버스 정류장 번호 리스트
bus_stop_numbers = [20810, 20169, 20172, 20535]

# 조건에 맞는 데이터 필터링
filtered_data = bus_station_data[bus_station_data['STTN_NO'].isin(bus_stop_numbers)]

# 중복 제거 (기준 날짜 무시하고 각 정류장 번호당 한 번만 보여주기)
unique_filtered_data = filtered_data.drop_duplicates(subset=['STTN_NO'])

# 필요한 정보만 선택하여 출력
print(unique_filtered_data[['STTN_NO', 'STTN_NM', 'CRDNT_X', 'CRDNT_Y']])

# 지도 시각화

import folium

# 버스 정류장 번호 리스트 및 이름 매핑
bus_stop_numbers = {
    20810: "숭실대입구역",
    20169: "숭실대정문앞",
    20172: "숭실대별관앞",
    20535: "숭실대후문"
}

# 버스 정류장 정보 리스트
bus_stations = [
    {"number": 20810, "name": "숭실대입구역", "lat": 37.496197, "lon": 126.9535564},
    {"number": 20169, "name": "숭실대정문앞", "lat": 37.495150, "lon": 126.957055},
    {"number": 20172, "name": "숭실대별관앞", "lat": 37.494985, "lon": 126.9599068},
    {"number": 20535, "name": "숭실대후문", "lat": 37.495095, "lon": 126.9610346}
]

# 버스 정류장별 이용 가능한 노선 정보 추가
bus_routes = {}
for stop_number, stop_name in bus_stop_numbers.items():
    available_lines = bus_user_data[bus_user_data['정류장명'].str.contains(stop_name)]['노선'].unique()
    bus_routes[stop_number] = available_lines

# 지도 생성
m = folium.Map(location=[37.495150, 126.957055], zoom_start=15)

# 정류장 마커 추가
for station in bus_stations:
    bus_stop_number = station['number']
    bus_list = bus_routes.get(bus_stop_number, [])
    popup_text = f"{station['name']}<br>이용 가능한 노선: {', '.join(bus_list)}"
    folium.Marker(
        location=[station['lat'], station['lon']],
        popup=popup_text,
        tooltip=station['name']
    ).add_to(m)

# 지도 출력
m
```
